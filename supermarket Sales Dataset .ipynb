```python
sns.countplot(x = 'Gender', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
sns.countplot(x = 'Product line', data = df)
plt.title('Product Line Distribution')
plt.xticks(rotation=90)
plt.show()
```

```python
# Partner status and Dependents
sns.countplot(x='Gender', hue='City', data=df)
plt.show()
```

```python
# Distribution of target column
sns.countplot(x=df['Product line'])
plt.xticks(rotation=90)
plt.show()
```

```python
# Stats summary of numerical data
df.describe()
```

```python
sns.heatmap(pd.crosstab(df['City'], df['Payment']))
plt.title('Product Line Distribution')
plt.xticks(rotation=90)
plt.show()
```

```python
df['Branch'].value_counts().plot(kind='bar')
```

```python
df['Payment'].value_counts().plot(kind='bar')
plt.title('Most Popular Payment')
plt.show()
```

```python
sns.boxplot(df['cogs'])
```

```python
sns.scatterplot(data=df, x='cogs', y='Rating')
```

```python

sns.barplot(x='Branch', y='gross income', data=df)
plt.title('The Most Profitable Branch As Per Gross Income')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Product line', y='Rating', data=df, palette="Set1")
plt.title('Customer Rating Distribution by Product Line', fontsize=16)
plt.savefig('Customer Rating Distribution by Product Line.png',dpi = 1000)
plt.show()
```

```python
sns.boxplot(x='Gender', y='gross income', data=df)
plt.title('The Relationship Between Gender And Gross Income')
plt.show()
```

```python
sns.heatmap(pd.crosstab(df['Gender'], df['City']))
plt.title('Product Line Distribution')
plt.xticks(rotation=90)
plt.show()
```
